````python
import pandas as pd

def process_weather(df_weather):
    """Function to structure and clean weather data.
    
    Structuring includes renaming the columns, selecting only 
    observations of maximum temperature and transforming temperature 
    from tenths of celcius to celcius. Cleaning includes sorting 
    and resetting the index.
    
    Args:
        df_weather (pd.DataFrame): DataFrame containing weather data
        
    Returns:
        pd.DataFrame: Processed weather data
    """
    # Structure and clean data using methods chaining
    df_out = (
        df_weather
            .rename(columns={'0': 'station', '1': 'datetime', '2': 'obs_type', '3': 'obs_value'})
            .query("obs_type == 'TMAX'")
            .assign(obs_value=lambda df: df['obs_value'] / 10)
            .sort_values(by=['station', 'datetime'], ascending=[True, False])
            .reset_index(drop=True)
            .copy()
    )

    # Area process
    df_out['area'] = df_out['station'].str[0:2]

    # Datetime process
    df_out['datetime_dt'] = pd.to_datetime(df_out['datetime'], format='%Y%m%d')
    df_out['month'] = df_out['datetime_dt'].dt.month
    df_out['year'] = df_out['datetime_dt'].dt.year

    return df_out

# Load the weather data
df_weather = pd.read_csv('data/weather_data_1870-1875.csv')

# Apply the process_weather function
df_weather_period = process_weather(df_weather)

# Display the resulting DataFrame
df_weather_period

####   Subquestion 3   ####
# The df_weather.iloc[:, :4] is not necessary since the dataframe loaded from the CSV file only has 4 columns. 
# Thereby the dataframe is not changed by the command.
````

> **Ex. 3.3.3.3:** Inspecting each line of the `lines` variable we see that the information about the column widths are all located from index `13` up and including index `17`. > Finally, apply `get_colspecs` to the `lines` variable and store the result in a new variable called `colspecs`.

````python
# Define function
def get_colspecs(lines):
    """Extracts colspecs from `ghcnd-stations-column-metadata.txt`.
    
    Args:
        lines (list[str]): 
            list of relevant rows from `ghcnd-stations-column-metadata.txt` 
    
    Returns:
        (list[str]): 
            list of extracted colspecs i.e. ['1-11', '13-20', ..., '81-85']
    """
    colspec_idx_start = 13
    colspec_idx_end = 17 + 1  # Including idx 17
    # Extract colspecs from each line
    colspecs = [line[colspec_idx_start:colspec_idx_end].strip() for line in lines]
    return colspecs

# Apply `get_colspecs` to the `lines` variable and store the result in a new variable called `colspecs`
colspecs = get_colspecs(lines)

# Print the colspecs to verify the result
print(colspecs)
````

> **Ex. 3.3.3.4:** Write a function named `get_colspec_pair` which takes as input a string variable named `colspec` and returns a tuple of integers. Specifically, the function should take a string similar to each element of `colspecs`, split this string by `-` and return a tuple of integers where
1. The first integer should have `1` subtracted from it (Python is 0-indexed!)
2. The second integer should stay as it is (the intervals provided to the pandas function `read_fwf` should be non-inclusive)
> As an example, applying the function to `"1-11"` and `"13-20"` should yield the following results:

```python
print(get_colspec_pair("1-11"))
## output: (0, 11)

print(get_colspec_pair("13-20"))
## output: (12, 20)
```


````python
def get_colspec_pair(colspec):
    """Converts a colspec string to a tuple of integers.
    
    Args:
        colspec (str): 
            String specifying the column range, e.g., '1-11'.
    
    Returns:
        (tuple): 
            Tuple of integers representing the start and end indices, adjusted for 0-based indexing, e.g (0, 11).
    """
    start, end = colspec.split('-')
    return (int(start) - 1, int(end))

# Example usage and verification
print(get_colspec_pair("1-11"))  # Expected output: (0, 11)
print(get_colspec_pair("13-20")) # Expected output: (12, 20)
````

````python
# Define function
def get_column_names(lines):
    """Extracts names from `ghcnd-stations-column-metadata.txt`.
    
    Args:
        lines (list[str]): 
            list of relevant rows from `ghcnd-stations-column-metadata.txt`
    
    Returns:
        (list[str]): 
            list of extracted names i.e. ['ID', 'LATITUDE', ...,]
    """
    colname_idx_start = 0
    colname_idx_end = 13
    
    # Extract column names from each line
    column_names = [line[colname_idx_start:colname_idx_end].strip() for line in lines]
    return column_names

# Apply `get_column_names` to the `lines` variable and store the result in a new variable called `column_names`
column_names = get_column_names(lines)

# Print the column names to verify the result
print(column_names)

# Define function
def get_column_dtypes(lines):
    """Extracts data types from `ghcnd-stations-column-metadata.txt`.
    
    Args:
        lines (list[str]): 
            list of relevant rows from `ghcnd-stations-column-metadata.txt`
    
    Returns:
        (list[str]): 
            list of extracted dtypes i.e. ['Character', 'Real', ...,]
    """
    dtype_idx_start = 18
    
    # Extract data types from each line
    column_dtypes = [line[dtype_idx_start:].strip() for line in lines]
    return column_dtypes

# Apply `get_column_dtypes` to the `lines` variable and store the result in a new variable called `column_dtypes`
column_dtypes = get_column_dtypes(lines)

# Print the column data types to verify the result
print(column_dtypes)
````

````python
#1.
# We compute the survival probability for each combination of passenger class and gender
surv_prob = titanic.groupby(['pclass', 'sex'])['survived'].mean().reset_index()

# We create a barplot showing the survival probability for each combination of passenger class and gender
fig, ax = plt.subplots()

sns.barplot(data=surv_prob,
            x='pclass',
            y='survived',
            hue='sex')

# We set the title, x- and ylabel and show the plot
ax.set_title('Survival Probability by Gender and Class')
ax.set_xlabel('Passenger Class')
ax.set_ylabel('Probability of Survival')

# Adjust layout to tight
plt.tight_layout()

# We display the plot
plt.show();
````
